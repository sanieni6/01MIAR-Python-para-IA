```python
# Your imports here...
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

## 01
Carga de datos de csv (valencia_pollution_dataset.csv) en un DataFrame

```python
path_csv = "valencia_pollution_dataset.csv"
csv_data = pd.read_csv(path_csv)
csv_data.head()
# csv_data.columns
```

## 02
- ¿Cuál es el rango temporal del dataset?
- Obtención del número de estaciones que tienen un sensor para medir monóxido de carbono

```python
# Rango temporal del dataset
print( 'Rango temporal del dataset: ', csv_data['fecha_carga'].min(), ' - ', csv_data['fecha_carga'].max())

# Numero de estaciones que tienen un sensor de medir monoxido de carbono
# Estaciones con sensor de monóxido de carbono (co no es NaN)
mask = csv_data['co'].notna()

# Eliminar duplicados por estación
estaciones_con_co = csv_data[mask].drop_duplicates(subset='nombre')

# Contar estaciones únicas
print("Número de estaciones que tienen un sensor de CO:", len(estaciones_con_co))
```

## 03

- Número de estación con la media más alta de mediciones de monóxido de carbono

```python
# Numero de estacion con la media mas alta de mediciones de monoxido de carbono

# Calcular la media de CO por estación identificada por objectid
co_means = csv_data.groupby('objectid')['co'].mean()

# Obtener el objectid con la media más alta
objectid_max_co = co_means.idxmax()

print("El numero de estacion con la media mas alta de CO es:", objectid_max_co)
```

## 04
Realizar el histograma de dióxido de nitrógeno de la estación del punto anterior

```python
# Filtrar los datos de la estacion con mayor media de CO
estacion_objetivo = csv_data[csv_data['objectid'] == objectid_max_co]

# Graficar historigrama
plt.figure(figsize=(10, 6))
plt.hist(estacion_objetivo['no2'].dropna(), bins=10, edgecolor='black')
plt.title(f'Histograma de NO2 - Estación con mayor media de CO')
plt.xlabel('Concentración de NO2')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

## 05
Realizar gráfica con  la línea temporal de las mediciones de ozono de la estación del punto anterior. Pista, la columna 'fecha_carga' habría que cambiarla de tipo.

```python
# Your solution here...
#  Cambiar el tipo de dato de la columna 'fecha_carga' a datetime
csv_data['fecha_carga'] = pd.to_datetime(csv_data['fecha_carga'])

# Filtrar datos de la estación con mayor media de CO
estacion_max_co = csv_data[csv_data['objectid'] == objectid_max_co]

# Ordenar por fecha para una mejor visualización
estacion_max_co = estacion_max_co.sort_values(by='fecha_carga')

# Graficar línea temporal de O3
plt.figure(figsize=(10, 5))
sns.lineplot(data=estacion_max_co, x='fecha_carga', y='o3', marker='o')
plt.title(f'Línea Temporal de Ozono (O3) - Estación')
plt.xlabel('Fecha de Carga')
plt.ylabel('Concentración de O3')
plt.grid(True)
plt.tight_layout()
plt.show()
```
