### Paso 1: Importar Librerías

```python
#import the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from glob import glob

# Configuraciones de estilo
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

### Paso 2: Cargar los datos

```python
# Ruta a la carpeta de datos (ajusta si tu path es distinto)
data_path = "./dati-regioni/"
csv_files = sorted(glob(os.path.join(data_path, "*.csv")))

# Cargar y concatenar todos los CSVs
dfs = [pd.read_csv(file) for file in csv_files]
df = pd.concat(dfs, ignore_index=True)

# Mostrar las primeras filas
df.head()
df.info()
```

```python
df.columns
```

### Paso 4: Limpieza y preprocesamiento básico

```python
# Revisar valores nulos
df.isnull().sum()
```

Seguiendo los pasos del whitepaper, no es necesario hacer ningun tipo de limpieza de datos, ya que, si bien existen campos con valores nulos, estos no se encuentran en los campos principales que usaremos para replicar la correlacion y las visualizaciones de paper. Estos no tienen valores nulos.

Procedemos a convertir la columna "data" a formato fecha.

```python
df['data'] = pd.to_datetime(df['data'])
```

Debido a que el dataset esta en Italiano, se procede a traducir los nombres de las columnas a ingles para una mejor comprension de los datos a evaluar.

```python
column_translation = {
    'data': 'date',
    'stato': 'state',
    'codice_regione': 'RegionCode',
    'denominazione_regione': 'regionName',
    'lat': 'latitude',
    'long': 'longitude',
    'ricoverati_con_sintomi': 'HospitalizedWithSymptoms',
    'terapia_intensiva': 'IntensiveCarePatients',
    'totale_ospedalizzati': 'TotalHospitalizedPatients',
    'isolamento_domiciliare': 'HomeConfinement',
    'totale_positivi': 'TotalPositiveCases',
    'variazione_totale_positivi': 'VariationTotalPositiveCases',
    'nuovi_positivi': 'NewPositiveCases',
    'dimessi_guariti': 'Recovered',
    'deceduti': 'Deaths',
    'totale_casi': 'TotalCases',
    'tamponi': 'TestsPerformed',
    'casi_testati': 'IndividualsTested'
}
```

Aplicar la traducción a una copia del DataFrame

```python
# Copiar el DataFrame original
df_en = df.copy()

# Renombrar columnas al inglés
df_en = df_en.rename(columns=column_translation)

# Mostrar para verificar
df_en[list(column_translation.values())]
```

### Paso 5: Generar matriz de correlación

```python
cols_to_correlate = [
    'TotalCases',
    'NewPositiveCases',
    'HospitalizedWithSymptoms',
    'IntensiveCarePatients',
    'TotalHospitalizedPatients',
    'HomeConfinement',
    'TotalPositiveCases',
    'VariationTotalPositiveCases',
    'Recovered',
    'Deaths',
    'TestsPerformed'
]

df_corr = df_en[cols_to_correlate]
```

```python
# Matriz de correlación
corr_matrix = df_corr.corr()

# Visualización con heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    center=0,
    vmin=-1,
    vmax=1
)
plt.title("Matriz de Correlación entre Variables COVID-19")
plt.show()
```

### Paso 6: Conteo total por variable

```python
# Selección de columnas relevantes para el análisis
selected_cols = [
    'TotalPositiveCases',
    'HomeConfinement',
    'TotalHospitalizedPatients',
    'IntensiveCarePatients',
    'Recovered',
    'Deaths',
    'TotalCases',
    'TestsPerformed',
]
```

```python
# Calcular los totales
totals_table = (
    df_en[list(selected_cols)]
    .sum()
    .reset_index()
    .rename(columns={'index': 'Detection', 0: 'Total'})
)

# Ordenar como en la imagen
totals_table = totals_table.sort_values(by='Total', ascending=False).reset_index(drop=True)
totals_table.index += 1  # Para que el índice inicie en 1

# Mostrar como tabla
totals_table
```

### Paso 7: Tabla por región ordenada por TotalPositiveCases

```python
# Agrupar por región y sumar las métricas relevantes
region_summary = (
    df_en.groupby('regionName')[[
        'TotalHospitalizedPatients',
        'Recovered',
        'Deaths',
        'TotalPositiveCases',
        'TestsPerformed'
    ]]
    .sum()
    .sort_values(by='TotalPositiveCases', ascending=False)
)

# Mostrar la tabla resultante
region_summary.reset_index()
```

### Paso 8: Tests realizados por región

```python
import matplotlib.ticker as mtick
# Agrupamos por región y sumamos el total de tests
tests_by_region = (
    df_en.groupby('regionName')['TestsPerformed']
    .sum()
    .sort_values(ascending=False)
)

# Convertimos a DataFrame para graficar
tests_df = tests_by_region.reset_index()

# Graficamos
plt.figure(figsize=(14, 6))
sns.barplot(data=tests_df, x='regionName', y='TestsPerformed', palette='tab10')

# Mostrar valores enteros con separador de miles
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Estética
plt.title("Tests Performed vs Region", fontsize=16)
plt.xlabel("RegionName")
plt.ylabel("TestsPerformed")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

- Lombardia encabeza claramente la lista, lo que está alineado con su rol como una de las regiones más afectadas por el COVID-19 en Italia.

### Paso 9: Tests por día



```python
# Agrupar por fecha y sumar tests realizados
tests_by_date = (
    df_en.groupby('date')['TestsPerformed']
    .sum()
    .reset_index()
)

# Graficar línea con sombreado (seaborn lineplot lo hace por defecto)
plt.figure(figsize=(12, 6))
sns.lineplot(data=tests_by_date, x='date', y='TestsPerformed')

plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.title("Tests Performed vs Date", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Tests Performed")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Paso 10: Lineplot de hospitalizados, recuperados y muertes

```python
# Agrupar por fecha y sumar cada variable
daily_totals = (
    df_en.groupby('date')[['TotalHospitalizedPatients', 'Recovered', 'Deaths']]
    .sum()
    .reset_index()
)

# Graficar las 3 líneas
plt.figure(figsize=(12, 6))

sns.lineplot(data=daily_totals, x='date', y='TotalHospitalizedPatients', label='Hospitalized Patients', color='black')
sns.lineplot(data=daily_totals, x='date', y='Recovered', label='Recovery', color='red')
sns.lineplot(data=daily_totals, x='date', y='Deaths', label='Deaths', color='blue')

plt.title("Hospitalized Patients, Recovery, Death vs Count", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

- Este gráfico ayuda a visualizar cómo la capacidad hospitalaria se va liberando conforme aumentan los recuperados.
- La intersección de las curvas sugiere un punto de cambio en la carga del sistema de salud.

### Paso 10: Gráfico apilado de tests vs casos confirmados

```python
# Agrupar por región y sumar tests y casos confirmados
tests_and_confirmed = (
    df_en.groupby('regionName')[['TestsPerformed', 'TotalPositiveCases']]
    .sum()
    .sort_values(by='TestsPerformed', ascending=False)
    .reset_index()
)

# Graficar barras apiladas
plt.figure(figsize=(14, 7))

# Barras de tests
plt.bar(tests_and_confirmed['regionName'], tests_and_confirmed['TestsPerformed'], label='Tests Performed', color='steelblue')

# Barras de casos confirmados encima (apilado)
plt.bar(tests_and_confirmed['regionName'], tests_and_confirmed['TotalPositiveCases'], label='Confirm Cases', color='orange')

plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))


plt.title("Test and Confirmed Cases vs Region", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

Regiones como Lombardia y Veneto tienen la mayor diferencia absoluta entre tests y confirmados, lo que puede reflejar:
- Buen nivel de tamizaje poblacional, o
- Alta carga viral pero también gran capacidad de pruebas.

Regiones como Molise o Basilicata tienen pocos tests y también pocos casos, probablemente por menor población.

### Paso 11: Scatterplot de Hospitalizados vs Positivos con color por Recuperados

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df_en,
    x='TotalPositiveCases',
    y='TotalHospitalizedPatients',
    hue='Recovered',
    palette='magma',
    edgecolor='none',
    alpha=0.9
)

plt.title("Total Hospitalized Patients vs Total Positive Cases", fontsize=14)
plt.xlabel("TotalPositiveCases")
plt.ylabel("TotalHospitalizedPatients")
plt.legend(title="Recovered", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- El patrón muestra que no todos los positivos requieren hospitalización.
- El gradiente de color sugiere que a medida que se acumulan más recuperados (color más oscuro), la relación entre casos positivos y hospitalizados se aplana o disminuye, posiblemente reflejando control o inmunidad creciente.

### Paso 13: Scatterplot de nuevos casos vs pacientes en UCI, coloreado por recuperados

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df_en,
    x='NewPositiveCases',
    y='IntensiveCarePatients',
    hue='Recovered',
    palette='magma',
    edgecolor='none',
    alpha=0.9
)

plt.title("New Positive Cases vs ICU Patients", fontsize=14)
plt.xlabel("NewPositiveCases")
plt.ylabel("IntensiveCarePatients")
plt.legend(title="Recovered", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- No todos los nuevos casos ingresan a UCI, lo que puede indicar variabilidad en la severidad del virus o buena capacidad de respuesta médica.
- Las zonas más oscuras (más recuperados) tienden a coincidir con mayor cantidad de nuevos casos, lo cual sugiere un sistema de salud más eficaz conforme avanzaba el brote.
